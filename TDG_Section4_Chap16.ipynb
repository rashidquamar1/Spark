# Configuring a spark application is essentially important before writting code

 Majority of configurations fall into the following categories:

 *  Application properties
 *  Runtime environment
 *  Shuffle behavior
 *  Spark UI
 *  Compression and serialization
 *  Memory management
 *  Execution behavior
 *  Networking
 *  Scheduling
 *  Dynamic allocation
 *  Security
 *  Encryption
 *  Spark SQL
 *  Spark streaming

**Application Properties**

*  spark.app.name=my-spark-app
*  spark.logConf=true


**Shuffle Behavior**

*  spark.sql.shuffle.partitions=200
*  spark.shuffle.compress=true
*  spark.shuffle.spill.compress=true
*  spark.reducer.maxSizeInFlight=96m

**Spark UI & Event Logs**

*  spark.ui.enabled=true
*  spark.eventLog.enabled=true
*  spark.eventLog.dir=dbfs:/spark-event-logs
*  spark.ui.retainedJobs=100
*  spark.ui.retainedStages=100


**Compression & Serialization**

*  spark.serializer=org.apache.spark.serializer.KryoSerializer
*  spark.io.compression.codec=lz4
*  spark.kryoserializer.buffer.max=512m

How Many Executors Per Node?

Rule of thumb:

25 executors per node

Avoid 1 huge executor

Avoid too many tiny executors

**Execution Behavior**

*  spark.task.maxFailures=4
*  spark.speculation=false

**Networking**

*  spark.network.timeout=600s
*  spark.rpc.askTimeout=120s
*  spark.shuffle.io.maxRetries=10
*  spark.shuffle.io.retryWait=5s

 Prevents executor loss on slow cloud storage


 Very important for ADLS/S3


**Dynamic Allocation**

*  spark.dynamicAllocation.enabled=true
*  spark.dynamicAllocation.minExecutors=2
*  spark.dynamicAllocation.initialExecutors=4
*  spark.dynamicAllocation.maxExecutors=50

Adjust maxExecutors based on:

maxExecutors  executorCores  cluster cores


# Sample Configuration

```python
spark.logConf=true
spark.executor.extraJavaOptions=-XX:+UseG1GC
spark.driver.extraJavaOptions=-XX:+UseG1GC

spark.sql.shuffle.partitions=200
spark.shuffle.compress=true
spark.shuffle.spill.compress=true

spark.serializer=org.apache.spark.serializer.KryoSerializer
spark.io.compression.codec=lz4

spark.executor.memory=8g
spark.executor.memoryOverhead=2g
spark.driver.memory=4g

spark.sql.adaptive.enabled=true
spark.sql.adaptive.coalescePartitions.enabled=true
spark.sql.adaptive.skewJoin.enabled=true

spark.network.timeout=600s

spark.scheduler.mode=FAIR

spark.dynamicAllocation.enabled=true
spark.dynamicAllocation.minExecutors=2
spark.dynamicAllocation.initialExecutors=4
spark.dynamicAllocation.maxExecutors=50
```
